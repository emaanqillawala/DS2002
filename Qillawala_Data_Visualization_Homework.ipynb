```python
# import libraries

from matplotlib import pyplot as plt

import numpy as np
import pandas as pd
```

### Import data through CSV as a dataframe

##### data source from Kaggle -> Top Grossing Movies Dataset
https://www.kaggle.com/datasets/akankshaaa013/top-grossing-movies-dataset

```python
# import data as a dataframe
movie_data = pd.read_csv('Top_Highest_Openings.csv', sep = ',')
print(movie_data.head())
movie_data.info()
```

### Graph 1: Box and Whisker Plot

```python
# box and whisker plot 
# showing the relationship between total gross from all of the movies released 
# by a particular movie distributor

# main box and whisker plot
boxandwhisker = movie_data.boxplot(grid = False, column = ['Total Gross'], by = ['Distributor'])
# added plot features
boxandwhisker.set_title('Top Gross for Movies by Movie Distributor')
plt.xticks(rotation='vertical')
boxandwhisker.set_xlabel('Movie Distributor')
boxandwhisker.set_ylabel('Movie Gross (in hundreds of millions of dollars)')
```

### Graph 2: Barplot

```python
# barplot
# showing the relationship between movie title and total number of theaters screening the release
# for only movies released in 2023 or later to show most accurate post-COVID data

# convert to datetime and index data frame by movie release date
movie_data['Date'] = pd.to_datetime(movie_data['Date'], format = '%d-%m-%Y')
recent_movie_data = movie_data.loc[movie_data.Date >= '01-01-2023', ]
recent_movie_data

# main barplot
fig = plt.figure(figsize=(8,6))
ax = plt.axes()

barplot = ax.bar(recent_movie_data['Release'], recent_movie_data['Theaters'], color = 'red', width = 0.25)
# added barplot features
ax.set_title('Number of Theaters Showing Each Movie Release for 2023 and Later')
plt.xticks(rotation='vertical', fontsize = 7)
ax.set_xlabel('Movie Release')
ax.set_ylabel('Number of Theaters Showing')
```

### Graph 3: Scatterplot

```python
# scatterplot 
# showing the relationship between total money earned during opening weekend 
# and the percent that open weekend earnings make up the total movie gross openings
# for only movies released in 2023 or later to show the most accurate post-COVID data

# main scatterplot
fig, ax = plt.subplots(figsize = (10, 8))

scatterplot = ax.scatter(recent_movie_data['Opening'], recent_movie_data['% of Total'], s = 200, marker = (5, 2), color = 'purple')
# added scatterplot features
ax.set_title('Opening Weekend Earnings and the Percentage it Represents The Movies Total Gross Earnings in 2023 or Later')
plt.xticks(rotation='vertical', fontsize = 10)
ax.set_xlabel('Total Opening Weekend Earnings (in hundreds of millions of dollars)')
ax.set_ylabel('Percent of Total Movie Gross (%)')
```
